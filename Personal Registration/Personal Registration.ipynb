```python
# CREATE
def create_person():
    try:
        national_code_sql = int(input('Enter person number: '))
    except: print ('Plaese write a number')
    first_name_sql = input('Enter first name: ')
    last_name_sql = input('Enter last name: ')
    post_code_sql = int(input('Enter postcode: '))
    gender_sql = input('Enter gender:')
    
    birth_date_sql = input('Enter date of birth: ')
    d1 = '1900-01-01'
    d2 = birth_date_sql
    d3 = datetime.today().strftime('%Y-%m-%d')

    d4 = datetime.strptime(d1, "%Y-%m-%d") <= datetime.strptime(d2, "%Y-%m-%d") <= datetime.strptime(d3, "%Y-%m-%d")
    if d4 == False :
        print(' Check Birth of date!')
    else :
        curs = conn.cursor()
        curs.execute("""INSERT INTO personal_data(national_code, first_name, last_name, address_id ,gender, birth_date) 
                  VALUES(%s, %s, %s, %s, %s, %s)""",(national_code_sql, first_name_sql, last_name_sql, post_code_sql, gender_sql, birth_date_sql))
        curs = conn.cursor()

        return print('You have added information of', first_name_sql,last_name_sql,'by national code:', national_code_sql, ', postcode:',post_code_sql,', gender:',gender_sql,'and Birth date:', birth_date_sql,'successfully!' )
```

```python
create_person()
```

```python
# CREATE
def create_address():
    try:
        post_code_sql = int(input('Enter a postcode: '))
    except: print ('Plaese write a number')
        
    city_sql = input('Enter a City name: ')
    street_sql = input('Enter a Street name: ')
    bld_no_sql = int(input('Enter Building No.: '))
    
    curs = conn.cursor()
    curs.execute("""INSERT INTO address(post_code, city, street, bld_no ) 
                  VALUES(%s, %s, %s, %s)""",(post_code_sql, city_sql, street_sql, bld_no_sql))
    curs = conn.cursor()

    return print('you have successfully added information of this address:\n Postcode:',post_code_sql,
                                                      '\n City:',city_sql,
                                                      '\n Street:',street_sql,
                                                        '\n NO.:',bld_no_sql)
```

```python
# UPDATE Persson
n=0
def update_person():
    try:
        national_code_sql = int(input('Enter person number: '))
    except: print ('Plaese write a number')
    try:
        n = int(input('''please choose 1-5 to update personal data:
                  \n 1-First Name 
                  \n 2-Last name 
                  \n 3-Postcode 
                  \n 4-Gender 
                  \n 5-Birth date \n'''))
    except:print ('Plaese write a number')
    if n==1:
        try:
            my_update = input('Enter first name: ')
        except: print ('Plaese write a name')
        curs = conn.cursor()
        curs.execute('''UPDATE personal_data SET first_name = %s  WHERE national_code = %s''',(my_update,national_code_sql))
        
    elif n==2:
        try:
            my_update = input('Enter last name: ')
        except: print ('Plaese write a name')
        curs = conn.cursor()
        curs.execute('''UPDATE personal_data SET last_name = %s  WHERE national_code = %s ''',(my_update,national_code_sql))
        
    elif n==3:
        try:
            my_update = int(input('Enter postcode: '))
        except: print ('Plaese write a number')
        curs = conn.cursor()
        curs.execute('''UPDATE personal_data SET address_id = %s  WHERE national_code = %s ''',(my_update,national_code_sql))
        
    elif n==4:
        
        my_update = input('Enter gender:')
        
        curs = conn.cursor()
        curs.execute('''UPDATE personal_data SET gender = %s  WHERE national_code = %s ''',(my_update,national_code_sql))
        
    elif n==5:
        my_update = input('Enter date of birth: ')
        d1 = '1900-01-01'
        d2 = my_update
        d3 = datetime.today().strftime('%Y-%m-%d')

        d4 = datetime.strptime(d1, "%Y-%m-%d") <= datetime.strptime(d2, "%Y-%m-%d") <= datetime.strptime(d3, "%Y-%m-%d")
        if d4 == False :
            raise Exception(print(' Check Birth of date!'))
        else :
            curs = conn.cursor()
            curs.execute('''UPDATE personal_data SET birth_date = %s  WHERE national_code = %s ''',(my_update,national_code_sql))
        
    else:
        print('Please choose 1-5')
        


    return print('You have updated',national_code_sql,'by',my_update)
```

```python
# UPDATE Address
n=0
def update_address():
    try:
        post_code_sql = int(input('Enter Postcode: '))
    except: print ('Plaese write a number')
    try:
        n = int(input('''please choose 1-3 to update Address data:
                  \n 1-City  
                  \n 2-Street 
                  \n 3-No. \n'''))
    except: print ('Plaese write a number')
    if n==1:
        try:
            my_update = input('Enter City name: ')
        except: print ('Plaese write a name')
        curs = conn.cursor()
        curs.execute('''UPDATE address SET city = %s  WHERE post_code = %s''',(my_update,post_code_sql))
    elif n==2:
        my_update = input('Enter Street name: ')
        curs = conn.cursor()
        curs.execute('''UPDATE address SET street = %s  WHERE post_code = %s ''',(my_update,post_code_sql))
    elif n==3:
        my_update = int(input('Enter No.: '))
        curs = conn.cursor()
        curs.execute('''UPDATE address SET bld_no = %s  WHERE post_code = %s ''',(my_update,post_code_sql))

    else:
        print('Please choose 1-3')
        


    return print('You have updated',post_code_sql,'by',my_update)
```

```python
# DELETE Person

def delete_person():
    try:
        national_code_sql = int(input('Enter person number(12 digits): '))
    except:
        print("Can't find your National code, Please insert 12 digit numbers")    
    curs = conn.cursor()
    curs.execute('''DELETE FROM personal_data WHERE national_code = %s ''',(national_code_sql))
   

    return print('You have deleted',national_code_sql)
```
